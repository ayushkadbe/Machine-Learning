```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier

# Load your dataset (replace 'test.csv' with your actual dataset file)
dataset = pd.read_csv(r"C:\Users\AYUSH\OneDrive\Desktop\Machine Learning\Assignment 17.ipynb\fitness_class_2212 - fitness_class_2212.csv")


# Select independent variables (features) and the target variable
X = dataset[['booking_id', 'months_as_member', 'weight']]
y = dataset['attended']

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

# Calculate mean for each column
mean_values = X_train.mean()

# Replace missing values with mean in numeric columns
X_train.fillna(mean_values, inplace=True)
X_test.fillna(mean_values, inplace=True)

# Standardize the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Create a Random Forest Classifier (you may need to adjust parameters)
classifier = RandomForestClassifier(n_estimators=100, random_state=0)

# Fit the classifier
classifier.fit(X_train_scaled, y_train)

# Make predictions on the test set and compare with actual values
y_pred = classifier.predict(X_test_scaled)
print(np.concatenate((y_pred.reshape(len(y_pred), 1), np.array(y_test).reshape(len(y_test), 1)), 1))
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC

# Load the dataset
dataset = pd.read_csv(r"C:\Users\AYUSH\OneDrive\Desktop\Machine Learning\Assignment 17.ipynb\fitness_class_2212 - fitness_class_2212.csv")

# Check for missing values
print("Missing values in the dataset:")
print(dataset.isnull().sum())

# Fill missing values with the mean of valid numeric columns
numeric_cols = dataset.select_dtypes(include=[np.number]).columns
dataset[numeric_cols] = dataset[numeric_cols].fillna(dataset[numeric_cols].mean())

# Extract features (X) and target variable (y)
X = dataset.iloc[:, [0, 1, 2]].values  # Selecting columns by index
y = dataset.iloc[:, -1].values

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

# Standardize the features
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

# Train the SVM classifier with a linear kernel
classifier = SVC(kernel='linear', random_state=0)
classifier.fit(X_train, y_train)

# Make predictions on a new data point
new_data_point = np.array([[17, 79.56, 8]])
predicted_attendance = classifier.predict(sc.transform(new_data_point))
print("Predicted Attendance:", predicted_attendance)

# Make predictions on the test set
y_pred = classifier.predict(X_test)

# Display the predicted and actual values side by side
result_comparison = np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test), 1)), axis=1)
print("Predicted vs Actual:")
print(result_comparison)
```
