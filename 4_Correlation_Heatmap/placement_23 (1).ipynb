```python
import numpy as np 
import pandas as pd
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# create dataframe
df=pd.read_csv(r'/home/infinitex/Desktop/JupiterNotebook/Assignment 4/placement.csv')
df
```

```python
df1 = df.drop(columns=['Gender','10th %'])
```

```python
df1.fillna(df.mean(numeric_only=True))
```

```python
df1['placed']='no'
df1.loc[df['specialisation'].notnull(), 'placed'] = 'yes'
```

```python
updated_csv_file_path = 'placementnew.csv'
df1.to_csv(updated_csv_file_path, index=False)
df1
```

```python
#encode columns Gender and placed uisng one hot encoding#
df2 = pd.get_dummies(df1, columns = ['12th Stream', 'Work exp'])
df2
```

```python
correlation_matrix =df2.corr(numeric_only =True)
correlation_matrix
```

```python
#display heatmap
plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.show()
```
