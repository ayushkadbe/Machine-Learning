# Area Plot

Area plots are similar to the line plot. They are also known as stack plots. These plots can be used to track changes over time for two or more related groups that make up one whole category. For example, lets compile the work done during a day into categories, say sleeping, eating, working and playing. Consider the below code.

```python
import matplotlib.pyplot as plt
days = [1,2,3,4,5]

sleeping =[7,8,6,11,7]
eating = [2,3,4,3,2]
working =[7,8,7,2,2]
playing = [8,5,7,8,13]
  
plt.plot([],[],color='m', label='Sleeping', linewidth=5)
plt.plot([],[],color='c', label='Eating', linewidth=5)
plt.plot([],[],color='r', label='Working', linewidth=5)
plt.plot([],[],color='k', label='Playing', linewidth=5)
  
plt.stackplot(days, sleeping,eating,working,playing, colors=['m','c','r','k'])
  
plt.xlabel('x')
plt.ylabel('y')
plt.title('Stack Plot')
plt.legend()
plt.show()
```

# Pie Chart

```python
import matplotlib.pyplot as plt
 
days = [1,2,3,4,5]
 
sleeping =[7,8,6,11,7]
eating = [2,3,4,3,2]
working =[7,8,7,2,2]
playing = [8,5,7,8,13]
slices = [7,2,2,13]
activities = ['sleeping','eating','working','playing']
cols = ['c','m','r','b']
 
plt.pie(slices,
  labels=activities,
  colors=cols,
  startangle=90,
  shadow= True,
  explode=(0,0.1,0,0),
  autopct='%1.1f%%')
 
plt.title('Pie Plot')
plt.show()
```

```python
import pandas as pd
df=pd.read_csv("full.csv")
```

# Create a bar plot to show the count of passengers by class

```python
class_counts = df['Pclass'].value_counts()
#print(class_counts)
plt.bar(class_counts.index, class_counts.values)
plt.title("Passenger Count by Class")
plt.xlabel("Class")
plt.ylabel("Count")
plt.xticks(class_counts.index)
plt.show()
```

# Create a histogram to show the distribution of ages

```python
plt.hist(df['Age'], bins=20, edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

# Create a pie chart to show the distribution of passengers by sex

```python
sex_counts = df['Sex'].value_counts()
plt.pie(sex_counts, labels=sex_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Passenger Distribution by Sex")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

# Create a scatter plot to show the relationship between age and fare

```python
plt.scatter(df['Age'], df['Fare'], alpha=0.5)
plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

# Count plot: Number of Survivors by Class (sns)

```python
import seaborn as sns
sns.set(style="whitegrid")
sns.countplot(data=df, x="Pclass", hue="Survived", palette="Set1")
plt.title("Number of Survivors by Class")
plt.xlabel("Class")
plt.ylabel("Count")
plt.legend(title="Survived", labels=["No", "Yes"])
plt.show()
```

# Box plot: Fare Distribution by Class

```python
sns.boxplot(data=df, x="Pclass", y="Fare", palette="Set2")
plt.title("Fare Distribution by Class")
plt.xlabel("Class")
plt.ylabel("Fare")
plt.show()
```

# Pair plot: Relationships between numerical variables

```python
numerical_features = ["Age", "Fare", "SibSp", "Parch"]
sns.pairplot(data=df, vars=numerical_features, hue="Survived", palette="Dark2")
plt.suptitle("Pair Plot of Numerical Features")
plt.show()
```
