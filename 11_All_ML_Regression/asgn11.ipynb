#3. Build simple linear regression


```python
X = wine.iloc[:,6:7].values
Y =wine.iloc[:,8:9].values
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test =train_test_split(X,Y, test_size=0.33,random_state=0)
X_train, X_test, Y_train, Y_test
```

```python
from sklearn.linear_model import LinearRegression
slrmodel = LinearRegression()
slrmodel.fit(X_train,Y_train)
Y_pred = slrmodel.predict(X_test)
Y_pred
```

#**4.Build multiple linear regression**

```python
from sklearn.metrics import mean_squared_error, r2_score
X = wine[['Total_Phenols','Proanthocyanins','OD280']]
Y = wine[['Flavanoids']]
X,Y
```

#Splitting train and test dataset

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
X_train, X_test, Y_train, Y_test
```

#building model and training model

```python
model = LinearRegression()
model.fit(X_train, Y_train)
```

#Predication and evalute the model

```python
Y_pred = model.predict(X_test)
mse = mean_squared_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")

print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
```

```python

plt.scatter(Y_test, Y_pred)
plt.xlabel("'Total_Phenols','Proanthocyanins','OD280'")
plt.ylabel("'Flavanoids'")
plt.title("Multiple linear regression")
plt.show()
```
