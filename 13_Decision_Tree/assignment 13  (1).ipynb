```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.tree import plot_tree
```

```python
dataset=pd.read_csv(r'/home/infinitex/Desktop/JupiterNotebook/Asgn 13 RandomForest/50_Startups - 50_Startups.csv")
X = dataset.iloc[:25, 0:3].values
Y= dataset.iloc[:25,4].values
X,Y
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state = 0)
regressor.fit(X, Y)
```

```python
regressor.predict([[100671.96,  91790.61, 249744.55]])
```

```python
plt.figure(figsize=(66,66))
plot_tree(regressor, filled=True)
plt.show()
```

```python
regressor = DecisionTreeRegressor(
    criterion='absolute_error',  # Mean Squared Error (mse) is the default criterion for regression tasks.
    max_depth=20,      # Set the maximum depth of the tree to 5.
    min_samples_split=6,  # Minimum samples required to split a node.
    min_samples_leaf=2,   # Minimum samples required at a leaf node.
    max_features=3,    # Consider all features for splitting.
    random_state=42      # Set a random seed for reproducibility.
)
regressor.fit(X,Y)
plt.figure(figsize=(66,66))
plot_tree(regressor, filled=True)
plt.show()
```
