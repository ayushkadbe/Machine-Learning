```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Assuming your dataset file is 'test.csv'
dataset = pd.read_csv(r"/home/infinitex/Desktop/JupiterNotebook/asgn 16/fitness_class_2212 - fitness_class_2212.csv")

# Convert 'days_before' to numeric by removing the ' days' suffix
dataset['days_before'] = dataset['days_before'].str.replace(' days', '').astype(float)

# Check for missing values
print("Missing values in the dataset:")
print(dataset.isnull().sum())
```

```python
# # Handle missing values by filling with the mean
# dataset.fillna(dataset.mean(), inplace=True)

X = dataset.iloc[:, [1, 2]].values  # Assuming 'months_as_member' is at position 1 and 'weight' is at position 2
y = dataset.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

print("X_train:")
print(X_train)
print("\ny_train:")
print(y_train)
print("\nX_test:")
print(X_test)
print("\ny_test:")
print(y_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

print("\nX_train (after scaling):")
print(X_train)
print("\nX_test (after scaling):")
print(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion='entropy', random_state=0)
classifier.fit(X_train, y_train)

# Example prediction
example_prediction = classifier.predict(sc.transform([[17, 79.56]]))
print("\nExample Prediction:")
print(example_prediction)

y_pred = classifier.predict(X_test)
results = np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test), 1)), axis=1)
print("\nComparison of Predictions and Actual Values:")
print(results)
```
